```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib as mpl
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
file = pd.read_csv('HT101_original.csv')
arquivo = file.loc[150:180]
arquivo
```

```python
fig = plt.figure(figsize=(8, 6))
t = fig.suptitle('Temp-Btm(F) - Temp-Brood(F)- Temp-Hive(F) ', fontsize=14)
ax = fig.add_subplot(111, projection='3d')

xs = list(file['Temp-Brood(F)'])
ys = list(file['Temp-Hive(F)'])
zs = list(file['HRH(%)'])

 #81.1 degrees Fahrenheit (27.3 degrees Celsius) in Louisiana and Texas to a low of 52.3 °F (11.3 °C) in Alaska.

#ax.scatter(xs, ys, zs, s=50, alpha=0.6, edgecolors='w')
#HRH(%) - concern: 90% and even 100% / normal:  40-60% // however it is
#the relative stability of the humidity rather than a precise value that emerges as a possible colony fitness indicator.


#temperature  - On the individual level, when temperature rises to more than 96.8 °F,
#honey bee brood is likely to be exposed to overheating.The normal range of temperature within a honey bee
#colony ranges from 91.4 to 96.8 °F //  Honey bees attempt to maintain this range as bee brood is very sensitive
#to changes in hive temperatures
#Honey bee colony losses can occur both during extreme cold and hot temperatures .colony losses occurred during severe 
#summer weather (temperature up to 114.8 °F) 

data_points = [(x, y, z) for x, y, z in zip(xs, ys, zs)]

ss = list(file['Weight(lbs)'])
colors = ['orange' if wt<= 5 else 'blue' if wt==9 else 'red' for wt in list(file['Month'])]
markers = ['x' if q>= 91.4 and q<=96.8 else ',' if q >96.8 else 'o' for q in list(file['Temp-Brood(F)'])]

for data, color, size, mark in zip(data_points, colors, ss, markers):
    x, y, z = data
    ax.scatter(x, y, z, alpha=0.4, c=color, edgecolors='none', s=size, marker=mark)

ax.set_xlabel('Temp-Brood(F)')
ax.set_ylabel('Temp-Hive(F)')
ax.set_zlabel('HRH(%)')
```
